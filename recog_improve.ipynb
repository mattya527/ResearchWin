```python
def number_of_elements(img):
    H = []
    S = []
    V = []
    for i in range(0, 32):
        for j in range(0, 32):
            H.append(img[i][j][0])
            S.append(img[i][j][1])
            V.append(img[i][j][2])
    return H, S, V
```
