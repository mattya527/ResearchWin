```python
def plot_loss_accuracy_graph(fit_record):
    plt.plot(fit_record.history['loss'],"-D",color="blue",label="train_loss",linewidth=2)
    plt.title('LOSS')
    plt.xlabel('EPOCHS')
    plt.ylabel('LOSS')
    plt.legend(loc="upper right")
    plt.show()
    
    plt.plot(fit_record.history['accuracy'],"-D",color="green",label="accuracy",linewidth=2)
    plt.title("ACCURACY")
    plt.xlabel('EPOCHS')
    plt.ylabel('ACCURACY')
    plt.legend(loc="upper right")
    plt.show()
```

```python
#CNNモデルの構築
model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Conv2D(16, (3, 3), activation='relu'))
# model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
# model.add(Dropout(0.5))
model.add(Dense(6,activation='softmax'))
model.summary()
```

```python
#モデルのコンパイル
model.compile(
    optimizer='adam',
    loss=categorical_crossentropy,
    metrics=['accuracy'],
)
```

```python
def plot_image(data_location,prediction_array,real_teacher_label,dataset):
    prediction_array,real_teacher_label,img = prediction_array[data_location],real_teacher_label[data_location],dataset[data_location]
    img = np.array(img,dtype='uint8')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    
    predicted_label = np.argmax(prediction_array)
    if predicted_label == real_teacher_label:
        color = "green"
    else:
        color = "red"
    plt.xlabel("{} {:2.0f}% ({})".format(classes[predicted_label], 100*np.max(prediction_array),classes[predicted_label]),color=color)
```

```python
def plot_teacher_labels_graph(data_location,prediction_array,real_teacher_label):
    prediction_array,real_teacher_label = prediction_array[data_location],real_teacher_label[data_location]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    thisplot = plt.bar(range(6),prediction_array,color="#666666")
    plt.xlim(0,1)
    predicted_label = np.argmax(prediction_array)
    thisplot[predicted_label].set_color('red')
    thisplot[real_teacher_label].set_color('green')
```

```python
def convertOneHotVector2Integers(one_hot_vector):
    return [np.where(r==1)[0][0] for r in one_hot_vector]
```
