```python
def plot_loss_accuracy_graph(fit_record):
    plt.plot(fit_record.history['loss'],"-D",color="blue",label="train_loss",linewidth=2)
    plt.title('LOSS')
    plt.xlabel('EPOCHS')
    plt.ylabel('LOSS')
    plt.legend(loc="upper right")
    plt.show()
    
    plt.plot(fit_record.history['accuracy'],"-D",color="green",label="accuracy",linewidth=2)
    plt.title("ACCURACY")
    plt.xlabel('EPOCHS')
    plt.ylabel('ACCURACY')
    plt.legend(loc="upper right")
    plt.show()
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
print(X.shape)
X = X.astype('float16')
reshape_size = (-1,32,32,1)
# X = X.reshape(reshape_size)
print(X.shape)
# X = X / 255.0
Y = to_categorical(Y,5)
#学習用とテスト用に分割
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,shuffle=True)
x_train = x_train.reshape(reshape_size)
print(x_train.shape,x_test.shape)
x_test = x_test.reshape(reshape_size)
```

```python
#CNNモデルの構築
model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(5,activation='softmax'))
model.summary()
```

```python
#モデルのコンパイル
model.compile(
    optimizer='adam',
    loss=categorical_crossentropy,
    metrics=['accuracy'],
)
```

```python
#学習
EPOCHS = 10
BATCH_SIZE = 16
print('反復学習回数:',EPOCHS)
fit_record = model.fit(
    x_train,
    y_train,
    batch_size=BATCH_SIZE,
    epochs = EPOCHS,
    verbose = 1,
)
plot_loss_accuracy_graph(fit_record)
```

```python
def plot_image(data_location,prediction_array,real_teacher_label,dataset):
    prediction_array,real_teacher_label,img = prediction_array[data_location],real_teacher_label[data_location],dataset[data_location]
    img = np.array(img,dtype='uint8')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    
    predicted_label = np.argmax(prediction_array)
    if predicted_label == real_teacher_label:
        color = "green"
    else:
        color = "red"
    plt.xlabel("{} {:2.0f}% ({})".format(classes[predicted_label], 100*np.max(prediction_array),classes[predicted_label]),color=color)
```

```python
def plot_teacher_labels_graph(data_location,prediction_array,real_teacher_label):
    prediction_array,real_teacher_label = prediction_array[data_location],real_teacher_label[data_location]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    thisplot = plt.bar(range(5),prediction_array,color="#666666")
    plt.xlim(0,1)
    predicted_label = np.argmax(prediction_array)
    thisplot[predicted_label].set_color('red')
    thisplot[real_teacher_label].set_color('green')
```
